```python
import numpy as np
from n2v.utils.evaluation_utils import best_PSNR
from matplotlib import pyplot as plt
```

```python
def compute_PSNR255(input_data, gt_data, fun):
    psnrs = []
    for img, gt in zip(input_data, gt_data):
        psnrs.append(fun(gt, img, 255.0))
    
    return np.array(psnrs)

def compute_PSNR(input_data, gt_data, fun):
    psnrs = []
    for img, gt in zip(input_data, gt_data):
        psnrs.append(fun(gt, img, gt.max() - gt.min()))
    
    return np.array(psnrs)

def plot_example(input_data, gt_data, vmin, vmax):
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.imshow(input_data, vmin=vmin, vmax=vmax, cmap="gray", interpolation="nearest")
    plt.axis('off');
    plt.subplot(1,2,2)
    plt.imshow(gt_data, vmin=vmin, vmax=vmax, cmap="gray", interpolation="nearest")
    plt.axis('off');
```

# BSD68

```python
test_x = np.load('../data/BSD68_reproducibility/test/bsd68_gaussian25.npy', allow_pickle=True)
test_x = [d[..., np.newaxis] for d in test_x]
test_gt = np.load('../data/BSD68_reproducibility/test/bsd68_groundtruth.npy', allow_pickle=True)
test_gt = [d[..., np.newaxis] for d in test_gt]
```

```python
scaled_PSNR = compute_PSNR255(test_x, test_gt, best_PSNR)
```

```python
print(f"Average input PSNR: {np.round(np.mean(scaled_PSNR), 2)}.")
```

```python
plot_example(test_x[0], test_gt[0], 0, 255)
```
